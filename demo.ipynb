```python
import numpy as np
from city_planner_gym import CityPlanningEnv
```

```python
# Initialize the environment
env = CityPlanningEnv()

# Reset the environment to get the initial state
obs = env.reset()
print("Initial Observation:")
print("Grid:\n", obs['grid'])
print("Budget:", obs['budget'])
print()

# Run a loop to perform random actions until the environment is done
done = False
step_count = 0
total_reward = 0
while not done:
    # Randomly select a row, column, and building type
    row = np.random.randint(0, 10)
    col = np.random.randint(0, 10)
    building_type = np.random.randint(0, 6)
    action = (row, col, building_type)

    # Take a step in the environment
    obs, reward, done, info = env.step(action)

    # Print the action and result
    print(f"Step {step_count}: Placing building type {building_type} at ({row}, {col})")
    print("Reward:", reward)
    print("Remaining Budget:", obs['budget'])
    print("Current Grid:\n", obs['grid'])
    print()

    # Increment step count
    step_count += 1
    total_reward += reward

print(f"Simulation finished. Total Reward: {total_reward}")
env.close()
```

```python
# Get average reward from playing randomly

env = CityPlanningEnv()

# Reset the environment to get the initial state

num_episodes = 5000
rewards = []
for i in range(num_episodes):
    done = False
    step_count = 0
    total_reward = 0
    obs = env.reset()
    while not done:
        # Randomly select a row, column, and building type
        row = np.random.randint(0, 10)
        col = np.random.randint(0, 10)
        building_type = np.random.randint(0, 6)
        action = (row, col, building_type)

        # Take a step in the environment
        obs, reward, done, info = env.step(action)

        # Increment step count
        step_count += 1
        total_reward += reward
    rewards.append(total_reward)

print(f'Average reward ({num_episodes} episodes): {sum(rewards) / len(rewards)}')
```
